__Forecasting stock prices__

Using moving average method

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math
from sklearn import metrics
```

```python
# reading from csv file
data1 = pd.read_csv(r"stock_info\infosys.csv")
data2 = pd.read_csv(r"stock_info\HEROMOTOCO.NS.csv")
data3 = pd.read_csv(r"stock_info\tatasteel.csv")
```

```python
break_point1 = 450
train1 = data1.iloc[:break_point1].copy()
test1 = data1.iloc[break_point1:].copy()

break_point2 = 450
train2 = data2.iloc[:break_point2].copy()
test2 = data2.iloc[break_point2:].copy()

break_point3 = 450
train3 = data3.iloc[:break_point3].copy()
test3 = data3.iloc[break_point3:].copy()
```

```python
len_of_moving_avg = 30

# for model1
predictions = []
train_data = list(train1['Adj Close'])
for i in range(len(test1)):
    temp = sum(train_data[-len_of_moving_avg:])/len_of_moving_avg
    train_data.append(temp)
    predictions.append(temp)
    
test1['movingAvgForecast'] = predictions

# for model2
predictions = []
train_data = list(train2['Adj Close'])
for i in range(len(test2)):
    temp = sum(train_data[-len_of_moving_avg:])/len_of_moving_avg
    train_data.append(temp)
    predictions.append(temp)
    
test2['movingAvgForecast'] = predictions

# for model3
predictions = []
train_data = list(train3['Adj Close'])
for i in range(len(test3)):
    temp = sum(train_data[-len_of_moving_avg:])/len_of_moving_avg
    train_data.append(temp)
    predictions.append(temp)
    
test3['movingAvgForecast'] = predictions
```

```python
# plotting outputs from model1
plt.figure()
plt.plot(train1['Adj Close'])
plt.plot(test1['Adj Close'])
plt.plot(test1['movingAvgForecast'])
plt.legend(['training', 'testing', 'predicted'])
plt.show()

# rms error
print("RMS error", math.sqrt(metrics.mean_squared_error(test1['Adj Close'], test1['movingAvgForecast'])))
```

```python
# plotting outputs from model2
plt.figure()
plt.plot(train2['Adj Close'])
plt.plot(test2['Adj Close'])
plt.plot(test2['movingAvgForecast'])
plt.legend(['training', 'testing', 'predicted'])
plt.show()

# rms error
print("RMS error", math.sqrt(metrics.mean_squared_error(test2['Adj Close'], test2['movingAvgForecast'])))
```

```python
# plotting outputs from model3
plt.figure()
plt.plot(train3['Adj Close'])
plt.plot(test3['Adj Close'])
plt.plot(test3['movingAvgForecast'])
plt.legend(['training', 'testing', 'predicted'])
plt.show()

# rms error
print("RMS error", math.sqrt(metrics.mean_squared_error(test3['Adj Close'], test3['movingAvgForecast'])))
```

__Conclusions:__

Moving average method is not a good forecasting method for stocks. It's value is going to settle down to a constant value after predicting enough number of times.
